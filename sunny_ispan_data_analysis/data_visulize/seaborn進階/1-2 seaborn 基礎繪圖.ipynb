## Seaborn
建立在 matplotlib 上的渲染神器! 你可以把他理解成很強的外掛

```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
```

## seaborn 和 pandas 有很好的整合性，可以直接匯入dataframe

```python
cars.info()
# mpg - 燃油效率（Miles Per Gallon）：每加侖汽油可以行駛的英里數。

# cylinders - 汽缸數：引擎中的汽缸數量，通常以數字表示，例如4、6、8。

# displacement - 排氣量：引擎的排氣容積，通常以立方英寸（cubic inches）或升（liters）表示。

# horsepower - 馬力：引擎的馬力數，衡量引擎的功率。

# weight - 重量：汽車的整體重量，通常以磅（pounds）為單位。

# acceleration - 加速度：汽車的加速性能，通常以秒鐘內達到60英里的速度來衡量。

# model_year - 車型年份：汽車的製造年份。

# origin - 產地：汽車的生產地，可能表示汽車的原產國家或地區。

# name - 名稱：汽車的型號或名稱。
```

```python
cars.dropna(inplace=True)
```

```python
cars.head()
```

## relplot

```python
sns.relplot(x="model_year", y="mpg", col="origin", hue="cylinders", data=cars)
```

## KDE plot
kernal density estimation

```python
cars.head()
```

```python
# 1維
sns.set_style("white")
sns.kdeplot(cars['horsepower'])
```

```python
sns.kdeplot(cars['horsepower'], fill=True)
```

```python
# 修改 bandwidth
sns.kdeplot(cars['horsepower'], bw_method = 0.01)
```

```python
sns.kdeplot(cars['horsepower'], bw_method = 10)
```

```python
# 累計
sns.kdeplot(cars['horsepower'], cumulative=True)
```

```python
# 2D
sns.set_style("whitegrid")
sns.kdeplot(x = cars['horsepower'], y = cars['mpg'])
```

```python
sns.kdeplot(x = cars['horsepower'], y = cars['mpg'], n_levels=20)
```

```python
sns.kdeplot(x = cars['horsepower'], y = cars['mpg'], fill=True)
```

```python
sns.set_style('whitegrid')

cyl_4 = cars[cars["cylinders"] == 4]
cyl_8 = cars[cars["cylinders"] == 8]

plt.figure(figsize=(8,6))

sns.kdeplot(x = cyl_4.horsepower, y = cyl_4.mpg,
            cmap="Blues", fill=True)
sns.kdeplot(x = cyl_8.horsepower, y = cyl_8.mpg,
            cmap="Reds", fill=True)

plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Miles per Gallon (MPG)', fontsize=14)

plt.annotate("4 Cylinders", (105, 32), color="blue", fontsize=16, fontweight='bold')
plt.annotate("8 Cylinders", (190, 18), color="red", fontsize=16, fontweight='bold')
```
